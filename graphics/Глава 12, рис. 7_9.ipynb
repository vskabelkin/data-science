```python
# -*- coding: utf-8 -*-
"""
 12, . 7_9
"""

import numpy as np
import random

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})

dim = 50
Y = np.where(np.random.rand(dim) > 0.5, dim, 0.0)
X = np.where(np.random.rand(dim) > 0.5, dim, 0.0)

xs = [random.random() for _ in range(dim)]  
ys = [random.random() for _ in range(dim)]  


def display_unit_vector():
    fig = plt.figure()
    
    plt.bar(xs,X,width=0.006,color='#348ABD')
    plt.gca().yaxis.set_major_locator(plt.NullLocator())
    plt.title(" ", fontsize=13)

    plt.show()
    #fig.savefig("fig_12-7", dpi=300, bbox_inches='tight', facecolor='w') 


def display_unit_square():
    fig = plt.figure()
    
    plt.scatter(xs, ys, marker='o', color='#348ABD')     # '#348ABD'
    plt.title(" ", fontsize=13)
    plt.axis([0,1,0,1])
    plt.show()
    #fig.savefig("fig_12-8", dpi=300, bbox_inches='tight', facecolor='w') 


def display_unit_cube():
    from mpl_toolkits.mplot3d import Axes3D
    def randrange(n, vmin, vmax):
        return (vmax - vmin)*np.random.rand(n) + vmin

    fig = plt.figure(figsize=(7,5), facecolor='w')
    ax = fig.gca(projection='3d',  axisbg='w')

    n = 100
    for c, m, zl, zh in [('#348ABD', 'o', 0, 1), ('#348ABD', 'o', 0, 1)]:
        xs = randrange(n, 0.0, 1.0)
        ys = randrange(n, 0.0, 1.0)
        zs = randrange(n, zl, zh)
        ax.scatter(xs, ys, zs, c=c, marker=m)

    plt.title(" \n", fontsize=13)
    plt.show()    
    #fig.savefig("fig_12-9", dpi=300, bbox_inches='tight', facecolor='w') #


display_unit_vector()
display_unit_square()
display_unit_cube()
```
