```python
# -*- coding: utf-8 -*-
"""
Глава 3, рис. 1
"""

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})

def make_chart_simple_line_chart():
    fig = plt.figure()

    years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
    gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

    # создать линейную диаграмму:, годы по оси X, ВВП по оси Y
    plt.plot(years, gdp, color='green', marker='.', linestyle='solid')

    # добавить название диаграммы
    plt.title("Номинальный ВВП", fontsize=13)

    # добавить подпись к оси Y
    plt.ylabel("Млрд. $", fontsize=11)
    plt.show()
    #fig.savefig("fig_3-1", dpi=300,bbox_inches='tight',facecolor='w') 

make_chart_simple_line_chart()
```
