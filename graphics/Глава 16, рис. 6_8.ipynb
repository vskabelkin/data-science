```python
# -*- coding: utf-8 -*-
"""
Глава 16, рис. 6_8
"""

import numpy as np
import math, random

from collections import Counter
from functools import partial, reduce

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})
    
def disp_matrix():  #6
    fig = plt.figure()
    
    def samplemat(d):
        mn = d
        """изготовить единичную матрицу (с единицами по диагонали)"""
        aa = np.zeros((d,d))

        for i in range(d):
            samp = random.sample(range(d),5)
            for j in samp:
                if j in range(i-2,i+2): continue
                aa[i,mn-j-1] = 1
        return aa  
    
    # Показать 2 матрицы разных размеров
    dim = 30

    a = samplemat(dim)

    xs = []
    for i in range(dim): 
        for j in range(dim):
            if a[i,j]==1 and i>dim-j-1: xs.append([i,dim-j])
    
    ys = []
    for i in range(dim): 
        for j in range(dim):
            if a[i,j]==1 and i<dim-j-1: ys.append([i,dim-j])

    y,x = list(zip(*xs))  
    y2,x2 = list(zip(*ys))  

    plt.scatter(x,y,marker='*', color="#348ABD")
    plt.scatter(x2,y2,marker='o', color="#348ABD")
    plt.plot([0,dim],[0,dim],'-', color="#348ABD")
    
    plt.xlim(-7,dim+8)
    plt.ylim(-5,dim+5)
    plt.title("Разделяющая гиперплоскость", fontsize=13)    
    plt.show()
    #fig.savefig("fig_16-6", dpi=300, bbox_inches='tight', facecolor='w') 

def example_set():   #7
    fig = plt.figure()
    
    xdata = [[-4,1,0],[-3,1,0],[-2,1,1],[-1,1,1],[0,1,1],[1,1,1],[2,1,1],[3,1,0],[4,1,0]]
    x = [[row[0],row[1]] for row in xdata if row[2]==0]
    x0, y0 = list(zip(*x))
    y = [[row[0],row[1]] for row in xdata if row[2]==1]
    x1, y1 = list(zip(*y))       
      
    plt.scatter(x1, y1, marker='o', label='положительный образцы', color="#348ABD")
    plt.scatter(x0, y0, marker='+', label='отрицательные образцы', color="#348ABD")
    
    plt.gca().yaxis.set_major_locator(plt.NullLocator())

    plt.xlim(-5,5)
    plt.legend(loc=9, fontsize=9)
    plt.show()     
    #fig.savefig("fig_16-7", dpi=300, bbox_inches='tight', facecolor='w')    
    
def separating_examples():   #8
    fig = plt.figure()
    
    xdata = [[-4,1,0],[-3,1,0],[-2,1,1],[-1,1,1],[0,1,1],[1,1,1],[2,1,1],[3,1,0],[4,1,0]]
    x = [[row[0],row[0]**2] for row in xdata if row[2]==1]
    x0, y0 = list(zip(*x))
    y = [[row[0],row[0]**2] for row in xdata if row[2]==0]
    x1, y1 = list(zip(*y))       
      
    plt.scatter(x0, y0, marker='o', label='положительный образцы', color="#348ABD")
    plt.scatter(x1, y1, marker='+', label='отрицательные образцы', color="#348ABD")
    plt.plot([-4,4],[6,6], color="#348ABD")

    plt.xlim(-5,5)
    plt.ylim(-5,20)
    plt.legend(loc=9, fontsize=9)
    plt.show()    
    #fig.savefig("fig_16-8", dpi=300, bbox_inches='tight', facecolor='w') 
     
    
disp_matrix()             #6
example_set()             #7
separating_examples()     #8

```
