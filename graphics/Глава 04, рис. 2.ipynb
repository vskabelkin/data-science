```python
# -*- coding: utf-8 -*-
"""
Глава 4, рис. 2
"""

import math #  математические функции

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})

def dot(v, w):
    """v_1 * w_1 + ... + v_n * w_n"""
    return sum(v_i * w_i for v_i, w_i in zip(v, w))
    
def scalar_multiply(c, v):
    return [c * v_i for v_i in v]

def make_graph_dot_product_as_vector_projection():
    fig = plt.figure()

    v = [2, 1]
    w = [math.sqrt(.25), math.sqrt(.75)]
    c = dot(v, w)
    vonw = scalar_multiply(c, w)
    o = [0,0]

    plt.arrow(0, 0, v[0], v[1], width=0.02, head_width=.1, length_includes_head=True)
    plt.annotate("v", v, xytext=[v[0] + 0.1, v[1]])
    plt.annotate("w", w, xytext=[w[0] - 0.15, w[1]])
    plt.arrow(0, 0, vonw[0], vonw[1], width=0.02, head_width=.05, length_includes_head=True)
    plt.annotate(u"(v•w)w", vonw, xytext=[vonw[0] - 0.1, vonw[1] + 0.1])
    plt.arrow(v[0], v[1], vonw[0] - v[0], vonw[1] - v[1], width=0.005, head_width=.05, linestyle=':', length_includes_head=True)
    
    plt.arrow(0 ,0, w[0], w[1], width=0.02, head_width=.1, length_includes_head=True)

    plt.scatter(*zip(v,w,o),marker='o', color='#348ABD')
    plt.axis([-0.5,2.5,-0.5,2.0])
    plt.title("Скалярное произведение", fontsize=13)
    plt.show()
    #fig.savefig("fig_4-2", dpi=300,bbox_inches='tight',facecolor='w')
  

make_graph_dot_product_as_vector_projection()    
```
