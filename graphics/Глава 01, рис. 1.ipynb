```python
# -*- coding: utf-8 -*-
"""
Глава 1, рис. 1
"""

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'}) # в Windows не работает  
mpl.rcParams.update({'font.family': 'Tahoma'})

salaries_and_tenures = [(83000, 8.7), (88000, 8.1), 
                        (48000, 0.7), (76000, 6),
                        (69000, 6.5), (76000, 7.5),
                        (60000, 2.5), (83000, 10),
                        (48000, 1.9), (63000, 4.2)]

def make_chart_salaries_by_tenure():
    fig = plt.figure()
    
    tenures = [tenure for salary, tenure in salaries_and_tenures]
    salaries = [salary for salary, tenure in salaries_and_tenures]
    
    plt.scatter(tenures, salaries)
    
    plt.xlabel("Стаж, лет", fontsize=11)     
    plt.ylabel("Заработная плата, доллары", fontsize=11)   
    plt.title("Зарплата в зависимости от опыта работы", fontsize=13)

    plt.show()  
    #fig.savefig("fig_1-1", dpi=300,bbox_inches='tight',facecolor='w') 

make_chart_salaries_by_tenure()
```
