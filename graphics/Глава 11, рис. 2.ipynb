```python
# -*- coding: utf-8 -*-
"""
Глава 11, рис. 2
"""
import numpy as np
import math, random

import warnings
warnings.simplefilter('ignore', np.RankWarning)

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})

def generate_nums(arr, n, d):
    return np.concatenate([np.linspace(x-x, x+(x/100*30), n) for x in arr])
    
def make_bestfit_polinamialsN(): 
    fig = plt.figure()
    
    x = [2.4, 2.8, 4.0, 4.3,  4.9,  5.3,  5.9,  7.9,  8.1,  8.7]
    y = [6.2, 7.6, 7.0, 7.5, 10.5, 10.7, 12.6, 15.0, 16.5, 17.5]
    z = np.polyfit(x, y, 9)     # коэффициенты
    p = np.poly1d(z)            # многочлены

    xr = generate_nums(x,10,2.4)
    yr = generate_nums(y,10,6.2)
    z2 = np.polyfit(xr, yr, 1)
    p2 = np.poly1d(z2)
   
    xr2 = generate_nums(x,100,2.4)
    yr2 = generate_nums(y,100,6.2)
    z3 = np.polyfit(xr2, yr2, 9)
    p3 = np.poly1d(z3)   
   
    xp = np.linspace(0, 20, 1000)    
    
    plt.plot(x, y, '*', marker='o', markersize=5)
    plt.plot(xp, p(xp), 'b-.',label='N=10')    
    plt.plot(xp, p2(xp), 'g--', label='N=100')    
    plt.plot(xp, p3(xp), 'r',  label='N=1000')

    plt.xlim(-1,11)
    plt.ylim(-1,21)
    plt.legend(loc=9,fontsize=9)
    plt.title("Подгонка полинома 9-ой степени с N точками", fontsize=13) 
    plt.show()
    #fig.savefig("fig_11-2", dpi=300, bbox_inches='tight', facecolor='w') 


make_bestfit_polinamialsN()  
```
