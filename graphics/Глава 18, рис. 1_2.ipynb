```python
# -*- coding: utf-8 -*-
"""
Глава 18, рис. 1_2
"""

import sys
sys.path.append("../../code-python3-ru")

import numpy as np

from lib.linear_algebra import dot
import math, random

import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline  

mpl.style.use('ggplot')  
mpl.style.use('seaborn-whitegrid')  
mpl.rcParams['axes.facecolor'] = u'fefefe'
#mpl.rcParams.update({'font.family': 'Ubuntu Condensed'})   
mpl.rcParams.update({'font.family': 'Tahoma'})


def decision_space():
    fig = plt.figure()
    
    x = [ 0, 0, 1, 1]
    y = [ 0, 1, 0, 1]
    labels = ['(0,0)', '(0,1)', '(1,0)', '(1,1)']

    plt.scatter(x, y)
    
    # присвоить метку каждой точке
    for label, x_count, y_count in zip(labels, x, y):
        plt.annotate(label,
                     xy=(x_count, y_count), # задать метку точке
                     xytext=(5, -5),        # и немного сместить ее
                     textcoords='offset points', color="#348ABD")    
    
    
    plt.plot([0,1.5],[1.5,0],'-',label='И-граница', color="#348ABD")
    plt.plot([-0.5,1],[1,-0.5],'g--',label='ИЛИ-граница')
    plt.axis([-0.5, 1.5, -0.5, 1.5])
    plt.xlabel("вход 1", fontsize=11) 
    plt.ylabel("вход 2", fontsize=11) 
    plt.legend(loc=1, fontsize=9)
    plt.title("Решающее пространство\nдля перцептрона с двумя входами", fontsize=13)
    plt.show()
    #fig.savefig("fig_18-1s", dpi=300, bbox_inches='tight', facecolor='w') 


def step_function(x):
    return 1 if x >= 0 else 0
def logistic(x):
    return 1.0 / (1 + math.exp(-x))
def sigmoid(t):
    return 1 / (1 + math.exp(-t))
def sigmoidv(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a    
    
def sigmoid_vs_step():
    fig = plt.figure()
    
    x = [-10, 0, 0, 10]
    x2 = np.arange(-10., 10., 0.2)
    y = [ 0, 0, 1,  1]    
    plt.plot(x,y,'b--',label='пороговая функция', color="#348ABD")   #, color="#348ABD"
    plt.plot(x2,sigmoidv(x2),'g',label='сигмоида')
    plt.axis([-10, 10, -0.1, 1.1])
    plt.legend(loc=2, fontsize=9)
    plt.title("Сигмоида и пороговая функция в сравнении", fontsize=13)
    plt.show()    
    #fig.savefig("fig_18-2s", dpi=300, bbox_inches='tight', facecolor='w') 
    
decision_space()
sigmoid_vs_step()
```
